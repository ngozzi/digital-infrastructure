```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append("../functions/")
from plot_style import plot_style
colors = plot_style("../functions/fonts/")
from multicollinearity import multicollinearity_diagnosis

# import data
df_col = pd.read_csv("./input-dfs-static/colombia_movs_std.csv")
df_ecu = pd.read_csv("./input-dfs-static/ecuador_movs_std.csv")
df_slv = pd.read_csv("./input-dfs-static/el-salvador_movs_std.csv")
```
